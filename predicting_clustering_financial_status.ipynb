#**ENCODING**

Clustering only accepts numerical variables. Therefore, we have to encode the categorical variables to
convert them into numerical variables.

```python
df_encoded = df.copy()
```

For k-means clustering, one-hot encoding is preferred for nominal categorical variables to avoid an
artificial ordinal relationship. However, for variables with many categories, label encoding may be more
practical to avoid a high-dimensional feature space

```python
df[categorical_columns].nunique()
```

From the code above, we can see that 'Market Cap(in B USD)' have over
100 unique values. As mentioned earlier, I will use label encoding for variables with a high number of
categories or unique values. I will set a threshold of 10 unique values as our standard.

*   One-Hot Encoding will be applied to categorical variables with less than 10 unique values.

*   Label Encoding will be applied to variables with more than 10 unique values.


```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
onehot_columns = []
label_encode_columns = []
for col in categorical_columns:
 if len(df[col].unique()) <= 10: onehot_columns.append(col)
 else: label_encode_columns.append(col)
```

```python
onehot_encoded = pd.get_dummies(df[onehot_columns])
```

```python
label_encoded = df[label_encode_columns].apply(LabelEncoder().fit_transform)
df_encoded = pd.concat([df.drop(categorical_columns, axis=1), onehot_encoded,
 label_encoded], axis=1)
```

#**STANDARDIZATION**

Now, the dataset is in a suitable format for k-means clustering, but I still need to scale the numerical
features. Scaling is crucial for k-means due to its reliance on distance calculations. We will use
StandardScaler, which standardizes features by removing the mean and scaling to unit variance. This is
particularly useful in preserving the shape of the dataset's distribution.

```python
df_scaled = df_encoded.copy()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled[numerical_columns] = scaler.fit_transform(df_scaled[numerical_columns])
```

#**MODELING CLUSTERING**


With the dataset now encoded and scaled, it is ready for clustering analysis, such as k-means. This
preprocessing ensures that the algorithm treats all features equally, which is especially important since
k-means is sensitive to the scale of the data.

I choose the k-means algorithm for clustering in this hotel booking demand dataset because of its
simplicity, scalability, and efficiency in identifying natural groupings in large datasets, making it suitable
for exploratory analysis and customer segmentation.

Before proceeding with clustering, I will determine the most optimal k value using two methods:
silhouette score and the elbow method. We will apply both methods to compare the results using both
df_encoded and df_scaled .


```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

#**ELBOW METHOD**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

K = range(2, 10)
distortions = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df_encoded)
    distortions.append(silhouette_score(df_encoded, kmeans.labels_))

plt.figure(figsize=(16, 8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('The Elbow Method for df_encoded')
plt.show()
```

From the silhouette score, the most optimal k value is 2, where the average silhouette score is 0.71.
Meanwhile, from the elbow method, the most optimal k value is 3, as it exhibits an elbow in the plot.

###**ANALYSIS**

From the proportions of clusters with a k value of 3 where the average silhouette score is 0.42, they are divided into 3 clusters: 8 data in cluster 0, 1 data in cluster 1, and 3 data in cluster 2.

In terms of revenue, the clusters can be categorized into three groups: lower to mid-tier, mid-tier, and upper mid-tier.

Cluster 0 is categorized within the lower to mid-tier because the companies within it are not yet stable. As seen in the table, the values in the Net Income and Earnings Per Share columns include negative values, indicating that these companies incurred losses. Cluster 0 likely comprises newly established businesses or those experiencing financial instability or uncertain performance in 2018.

Cluster 1 is categorized into the mid-tier. Looking at the table, the revenue for each company in this cluster is not as low as in cluster 0 but also not as high as in cluster 2. Despite having only one data, this cluster represents a mid-tier company and is not considered an outlier because the company in cluster 1 has the highest number of employees among the three clusters.

Cluster 2 is categorized within the upper mid-tier. The data in the revenue and net income columns for each company in this cluster are significantly higher than those in the other two clusters. It can be considered the best of the best, indicating stability as there are no negative values found in this table. This cluster is labeled as the top-performing cluster due to its consistently high values.

This analysis suggests that the clusters are distributed across different tiers based on their financial performance, stability, and key indicators such as revenue, net income, and employee count.
